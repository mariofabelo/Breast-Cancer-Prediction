## Preprocessing - One-hot encoding of categorical feature (diagnostic)

```python
%pip install scikit_learn
from sklearn.preprocessing import LabelEncoder

lb=LabelEncoder()

df.iloc[:, 1] = lb.fit_transform(df.iloc[:, 1].values)
df.head(10)
```

## Use Pairplot to observe possible optimal algorithms

```python
sns.pairplot(df.iloc[:, 1:5], hue="diagnosis")
```

```python
# dependent and independent features

X = df.iloc[:, 2:32].values # start at column 2 as 0 and 1 are not data
print(X)
y = df.iloc[:, 1].values  # diagnosis column
print(y)
```

### Splitting into training and test sets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### Standardise data for training (-1 to 1)

```python
from sklearn.preprocessing import StandardScaler

st = StandardScaler()

X_train = st.fit_transform(X_train)
print("X_train:", X_train)
X_test = st.fit_transform(X_test)
print("\n X_test:", X_test)
```

```python
X_train.shape
```

```python
y_train.shape

y_train
```

## Machine Learning Algorithm (Linear & Logistic Regression)

```python
from sklearn.linear_model import LogisticRegression, LinearRegression

log = LinearRegression()
```

```python
log.fit(X_train, y_train)
```

```python
log.score(X_train, y_train)
```

```python
log = LogisticRegression()

# Convert y from object to integer type for LogisticRegression
y_train = y_train.astype(int)


log.fit(X_train, y_train)
log.score(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# change y_test from object to integer type
y_test = y_test.astype(int)

accuracy_score(y_test, log.predict(X_test))
```

```python
print(classification_report(y_test, log.predict(X_test)))
```
