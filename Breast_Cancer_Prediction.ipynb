## Plot some decision boundaries

```python
def plot_decision_boundary(X, y, log, x_axis, y_axis):
  """
  Plots the decision boundary of the logistic regression model for the first two features.

  Args:
      X: The data matrix.
      y: The target labels.
      log: The trained logistic regression model.
      feature_names: Names of the features (optional).
  """

  # Create a meshgrid
  X1 = X[:, 0]
  X2 = X[:, 1]
  x_min, x_max = X1.min() - 1, X1.max() + 1
  y_min, y_max = X2.min() - 1, X2.max() + 1
  xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

  # Create a meshgrid of points for prediction
  mesh_points = np.c_[xx.ravel(), yy.ravel()] 

  # Scale the meshgrid points using the same scaler
  mesh_points_scaled = st.transform(mesh_points)

  # Make predictions on the scaled grid points
  Z = log.predict_proba(mesh_points_scaled)[:, 1]
  Z = Z.reshape(xx.shape)

  # Plot the decision boundary
  plt.contourf(xx, yy, Z, alpha=0.3)

  # Plot the data points with labels
  #plt.scatter(X[:, 0], X[:, 1], c=y) 

  # Plot the data points
  colors = ['purple', 'yellow']  # Define colors for each class
  diagnostics = ['benign', 'malignant']
  handles = []

  for i, diagnostic in enumerate(diagnostics):
      mask = y == diagnostic
      handles.append(plt.scatter([], [], c=colors[i], label = f'{diagnostic}'))

  plt.scatter(X[:, 0], X[:, 1], c=y)  
  plt.title('Logistic Regression Decision Boundary')
  plt.xlabel(x_axis)
  plt.ylabel(y_axis)

  plt.legend(handles=handles)

  plt.show()
```

### Decision boundary for radius mean and texture mean

```python
# Select the desired features (radius_mean and texture_mean)
X_desired = df.iloc[:, 2:4].values # only consider data from 3rd and 4th columns
print(X_desired)

# same y (diagnosis) as before

X_train, X_test, y_train, y_test = train_test_split(X_desired, y, test_size=0.2, random_state=0)

X_train_scaled = st.fit_transform(X_train)
print("X_train:", X_train_scaled)
X_test_scaled = st.fit_transform(X_test)
print("\n X_test:", X_test_scaled)

X_train_scaled.shape
y_train.shape
y_train = y_train.astype(int)

log.fit(X_train_scaled, y_train)

# Inverse transform the scaled data
X_train_original = st.inverse_transform(X_train_scaled) 
```

```python
plot_decision_boundary(X_train_original, y_train, log, x_axis = "radius_mean", y_axis = "texture_mean")
```
